# Write a python program to replicate a Banking system. The following features are mandatory:
1. Account login
2. Amount Depositing
3. Amount Withdrawal
Other than the above features you can add any other also.

```python
class MyAccount:
    def __init__(self, account_number, pin, balance=0):
        self.account_number = account_number
        self.pin = pin
        self.balance = balance

    def login(self, account_number, pin):
        if account_number == self.account_number and pin == self.pin:
            print("Login successful!")
            return True
        else:
            print("Invalid account number or pin. Please try again.")
            return False

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("Amount deposited successfully.")
            self.display_balance()
        else:
            print("Invalid amount. Please enter a positive value.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print("Amount withdrawn successfully.")
            self.display_balance()
        else:
            print("Insufficient balance or invalid amount.")

    def display_balance(self):
        print("Current Balance:", self.balance)


def main():
    # Create a new bank account
    account_number = input("Enter account number: ")
    pin = input("Set your 4-digit pin: ")
    myaccount = MyAccount(account_number, pin)

    # Login
    while True:
        entered_account_number = input("Enter your account number: ")
        entered_pin = input("Enter your 4-digit pin: ")
        if myaccount.login(entered_account_number, entered_pin):
            break

    # Banking operations
    while True:
        print("\nChoose operation:")
        print("1. Withdrawal")
        print("2. Deposit")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter choice (1/2/3/4): ")

        if choice == '1':
            amount = float(input("Enter amount to withdraw: "))
            myaccount.withdraw(amount)
        elif choice == '2':
            amount = float(input("Enter amount to deposit: "))
            myaccount.deposit(amount)
        elif choice == '3':
            myaccount.display_balance()
        elif choice == '4':
            print("Thank you for being a wonderful customer. Have a nice day!")
            break
        else:
            print("Invalid choice. Please enter a valid option.")


if __name__ == "__main__":
    main()
```
